```python
import pandas as pd

mixed_cols = []

for col in df1.columns:
    # ê²°ì¸¡ì¹˜ëŠ” ì œì™¸í•˜ê³ , ë¬¸ìì—´ì´ í•˜ë‚˜ë¼ë„ í¬í•¨ë˜ì–´ ìˆëŠ”ì§€ í™•ì¸
    sample = df1[col].dropna().astype(str).head(100)  # ìƒ˜í”Œ 100ê°œë§Œ ê²€ì‚¬
    # ìˆ«ìë§Œìœ¼ë¡œ ì´ë£¨ì–´ì§„ ê°’ì¸ì§€ ì²´í¬
    has_non_numeric = sample.apply(lambda x: not x.isdigit()).any()
    if has_non_numeric:
        mixed_cols.append(col)

print("ìˆ«ìì™€ ë¬¸ìê°€ ì„ì—¬ ìˆì„ ê°€ëŠ¥ì„±ì´ ìˆëŠ” ì»¬ëŸ¼:")
for c in mixed_cols:
    print(c)
```

```python
import re

def keep_only_numbers(x):
    # ê²°ì¸¡ì¹˜ëŠ” ê·¸ëŒ€ë¡œ ìœ ì§€
    if pd.isna(x):
        return np.nan
    # ë¬¸ìì—´ë¡œ ë³€í™˜ í›„ ìˆ«ìë§Œ ì¶”ì¶œ
    s = str(x)
    numbers = re.findall(r'\d+', s)  # ì—°ì†ëœ ìˆ«ì ë©ì–´ë¦¬ ì¶”ì¶œ
    if numbers:
        # ìˆ«ì ë©ì–´ë¦¬ë¥¼ ëª¨ë‘ ì´ì–´ë¶™ì—¬ ì •ìˆ˜í˜•ìœ¼ë¡œ ë³€í™˜
        return int(''.join(numbers))
    else:
        # ìˆ«ìê°€ ì „í˜€ ì—†ìœ¼ë©´ NaN ë°˜í™˜
        return np.nan

# ìˆ«ì+ë¬¸ì ì„ì¸ ì»¬ëŸ¼ ë¦¬ìŠ¤íŠ¸
cols_to_clean = [
    'ìº í˜ì¸ì ‘ì´‰ê±´ìˆ˜_R12M',
    'ìº í˜ì¸ì ‘ì´‰ì¼ìˆ˜_R12M'
]

# ê° ì»¬ëŸ¼ì— ëŒ€í•´ ìˆ«ìë§Œ ë‚¨ê¸°ë„ë¡ ë³€í™˜ ìˆ˜í–‰
for col in cols_to_clean:
    df1[col] = df1[col].apply(keep_only_numbers)
```

```python
# ìˆ«ìê°€ ë“¤ì–´ê°„ ì»¬ëŸ¼ë§Œ ì„ íƒ
numeric_cols = df1.select_dtypes(include=['number']).columns

# ê° ì»¬ëŸ¼ì˜ ê°’ë³„ ê°œìˆ˜ ì „ì²´ ì¶œë ¥
for col in numeric_cols:
    print(f"\nğŸ”¢ [ì»¬ëŸ¼: {col}] ê°’ë³„ ë¹ˆë„ìˆ˜ (ì „ì²´)")
    print(df1[col].value_counts(dropna=False).sort_index())
```

```python
# ID ì œì™¸í•œ ì»¬ëŸ¼ ì„ íƒ
num_cols = df1.drop(columns=['ID']).select_dtypes(include='number').columns

outlier_info = []

for col in num_cols:
    # ì´ìƒì¹˜ ê¸°ì¤€ ê³„ì‚° (IQR ë°©ì‹)
    q1 = df1[col].quantile(0.25)
    q3 = df1[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    # ì´ìƒì¹˜ ê°œìˆ˜ ê³„ì‚°
    outlier_count = df1[(df1[col] < lower) | (df1[col] > upper)].shape[0]
    
    if outlier_count > 0:
        outlier_info.append({
            'ì»¬ëŸ¼ëª…': col,
            'ì´ìƒì¹˜ ìˆ˜': outlier_count,
            'í•˜í•œê°’': lower,
            'ìƒí•œê°’': upper
        })

# ê²°ê³¼ ì •ë¦¬
outlier_df = pd.DataFrame(outlier_info).sort_values(by='ì´ìƒì¹˜ ìˆ˜', ascending=False).reset_index(drop=True)

print("ğŸ“Œ ì´ìƒì¹˜ê°€ ìˆëŠ” ì»¬ëŸ¼:")
display(outlier_df)
```

```python
import math

# ì´ìƒì¹˜ê°€ ìˆëŠ” ì»¬ëŸ¼ëª… ë¦¬ìŠ¤íŠ¸ë§Œ ì¶”ì¶œ (ìœ„ì—ì„œ ë§Œë“  outlier_df í™œìš©)
cols_with_outliers = outlier_df['ì»¬ëŸ¼ëª…'].tolist()

# 1í–‰ì— 5ê°œì”© ë°°ì¹˜
n_cols = 5
n_rows = math.ceil(len(cols_with_outliers) / n_cols)

# í”Œë¡¯ í¬ê¸° ì„¤ì •
plt.figure(figsize=(n_cols * 4, n_rows * 4))

# ë°•ìŠ¤í”Œë¡¯ ê·¸ë¦¬ê¸°
for i, col in enumerate(cols_with_outliers):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.boxplot(data=df1, y=col)
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
import pandas as pd

# 1) ID, OSêµ¬ë¶„ì½”ë“œ, ê·¸ë¦¬ê³  ë‚ ì§œí˜•(ìˆ«ìí˜•ìœ¼ë¡œ ì €ì¥ëœ) ì»¬ëŸ¼ì„ ëª¨ë‘ ì œì™¸í•œ ìˆ˜ì¹˜í˜• ì»¬ëŸ¼ëª… ë¦¬ìŠ¤íŠ¸ ìƒì„±
date_cols = ['ê¸°ì¤€ë…„ì›”']  

num_cols = (
    df1
    .drop(columns=['ID'] + date_cols)     # ì œì™¸í•  ì»¬ëŸ¼ë“¤ ì§€ì •
    .select_dtypes(include='number')                   # ìˆ«ìí˜• íƒ€ì…ë§Œ ì„ íƒ
    .columns
)

# 2) ì´ìƒì¹˜ í†µê³„ë¥¼ ë‹´ì„ ë¦¬ìŠ¤íŠ¸ ì´ˆê¸°í™”
outlier_stats = []

for col in num_cols:
    # 1ì‚¬ë¶„ìœ„(Q1)ì™€ 3ì‚¬ë¶„ìœ„(Q3)ë¥¼ ê³„ì‚°
    Q1, Q3 = df1[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    
    # ì´ìƒì¹˜ íŒì • ê¸°ì¤€: í•˜í•œ(lower), ìƒí•œ(upper)
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # ì´ìƒì¹˜ ë§ˆìŠ¤í¬, ê°œìˆ˜, ë¹„ìœ¨ ê³„ì‚°
    mask = (df1[col] < lower) | (df1[col] > upper)
    count = mask.sum()
    pct   = mask.mean() * 100
    
    # ì´ìƒì¹˜ê°€ 1ê°œ ì´ìƒì¸ ê²½ìš°ì—ë§Œ ì €ì¥
    if count > 0:
        outlier_stats.append((col, count, pct))

# 3) ê²°ê³¼ë¥¼ DataFrameìœ¼ë¡œ ë³€í™˜í•˜ê³ , ì´ìƒì¹˜ ë¹„ìœ¨ ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬ í›„ ì¸ë±ìŠ¤ ë¦¬ì…‹
outlier_df = (
    pd.DataFrame(outlier_stats, columns=['ì»¬ëŸ¼ëª…', 'ì´ìƒì¹˜ ê°œìˆ˜', 'ì´ìƒì¹˜ ë¹„ìœ¨(%)'])
      .sort_values(by='ì´ìƒì¹˜ ë¹„ìœ¨(%)', ascending=False)
      .reset_index(drop=True)
)

# 4) pandas ì¶œë ¥ ì˜µì…˜ ë³€ê²½: ëª¨ë“  í–‰Â·ì—´ì„ í‘œì‹œ
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# 5) ê²°ê³¼ ì¶œë ¥
print("ğŸ“Œ ìˆ˜ì¹˜í˜•(ë‚ ì§œí˜• ì œì™¸) ì»¬ëŸ¼ë³„ ì´ìƒì¹˜ í†µê³„:")
display(outlier_df)
```
