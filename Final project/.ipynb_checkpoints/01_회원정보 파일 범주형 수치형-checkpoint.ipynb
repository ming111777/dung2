```python
# ê¸°ë³¸ ë¼ì´ë¸ŒëŸ¬ë¦¬
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

# ê·¸ë˜í”„ ê¸°ë³¸ ì„¤ì •
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

# ê²½ê³  ëœ¨ì§€ ì•Šê²Œ
import warnings
warnings.filterwarnings('ignore')
```

```python
# parquet íŒŒì¼ ë¶ˆëŸ¬ì˜¤ê¸°
df = pd.read_parquet ('open/train/1.íšŒì›ì •ë³´/2018_train_íšŒì›ì •ë³´.parquet')
df
```

```python
# 1. ë‚ ì§œ ì»¬ëŸ¼ ì²˜ë¦¬
date_cols = [
    'ê¸°ì¤€ë…„ì›”', 'ì…íšŒì¼ì_ì‹ ìš©', 'ìµœì¢…ìœ íš¨ë…„ì›”_ì‹ ìš©_ì´ìš©ê°€ëŠ¥',
    'ìµœì¢…ìœ íš¨ë…„ì›”_ì‹ ìš©_ì´ìš©', 'ìµœì¢…ì¹´ë“œë°œê¸‰ì¼ì'
]

for col in date_cols:
    # ì´ìƒì¹˜(ì˜ˆ: 0, 999999 ë“±) ë¨¼ì € ì œê±°
    df[col] = df[col].replace([0, 999999, 99999999], pd.NA)
    # YYYYMM í˜•ì‹ì´ë¼ê³  ê°€ì •
    df[col] = pd.to_datetime(df[col].astype(str), format='%Y%m', errors='coerce')
```

```python
# 2. ìˆ˜ì¹˜í˜•/ë²”ì£¼í˜• ì»¬ëŸ¼ ë¶„ë¦¬
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

# ìˆ˜ì¹˜í˜•ì´ì§€ë§Œ ë²”ì£¼í˜•ì²˜ëŸ¼ ë³´ì´ëŠ” ì»¬ëŸ¼ë“¤ ì°¾ê¸° (ê³ ìœ ê°’ì´ ì ì€ ì •ìˆ˜í˜•)
numeric_to_categorical = []
for col in numeric_cols:
    unique_vals = df[col].dropna().unique()
    if pd.api.types.is_integer_dtype(df[col]) and len(unique_vals) <= 10:
        numeric_to_categorical.append(col)

# ì¡°ì •
numeric_cols = [col for col in numeric_cols if col not in numeric_to_categorical]
categorical_cols += numeric_to_categorical
```

```python
# 3. ìˆ˜ì¹˜í˜• / ë²”ì£¼í˜• ë°ì´í„°í”„ë ˆì„ ë¶„ë¦¬
df_numeric = df[numeric_cols].copy()
df_categorical = df[categorical_cols].copy()
```

```python
# 4. ë²”ì£¼í˜• ë³€ìˆ˜ ê³ ìœ ê°’ ìš”ì•½
print("ğŸ¯ ë²”ì£¼í˜• ë³€ìˆ˜ ê³ ìœ ê°’ ë¶„í¬:")
for col in df_categorical.columns:
    print(f"\nğŸ”¸ {col} (ê³ ìœ ê°’ {df_categorical[col].nunique()}ê°œ):")
    print(df_categorical[col].value_counts(dropna=False).head(10))  # ìƒìœ„ 10ê°œê¹Œì§€ë§Œ ë³´ê¸°
```
