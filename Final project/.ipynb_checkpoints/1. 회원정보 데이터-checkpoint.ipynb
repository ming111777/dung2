### 결측치 확인

```python
# 결측치가 있는 컬럼을 확인한다.

na_box = []

for idx, val in df1.isnull().sum().items() :
    print(f'{idx}의 결측치 : {val}')
    if val > 0 :
        na_box.append(idx)
```

### 6_회원여부_이용가능_CA

```python
df1['회원여부_이용가능_CA'].value_counts()
```

```python
df1['회원여부_이용가능_CA'].value_counts()/df1.shape[0] 
```

- 현금서비스 이용이 불가능한 회원의 수는 44944 (전체의 약 11%에 해당하는 수)

```python
# 블랙리스트인 회원 중 현금 서비스 이용이 가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(cond1 & cond2)
```

```python
# 블랙리스트가 아닌 회원 중 현금 서비스 이용이 가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(~cond1 & cond2)
```

```python
# 블랙리스트인 회원 중 현급 서비스 이용이 불가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(cond1 & ~cond2)
```

```python
# 블랙리스트가 아닌 회원 중 현급 서비스 이용이 불가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(~cond1 & ~cond2)
```
