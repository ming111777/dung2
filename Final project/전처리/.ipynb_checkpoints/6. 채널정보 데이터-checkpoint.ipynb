### 2_인입횟수_ARS_R6M
- 최근 6개월간 고객이 ARS에 전화한 횟수

```python
df1['인입횟수_ARS_R6M'].value_counts()
```

```python
# 문자열 제거하고 숫자만 추출해서 새로운 컬럼 만들기 (예: '10회 이상' → 10)
# 만약 원본이 이미 깨져 있다면, 아래처럼 정제해야 함

# 예시 (전처리된 컬럼이라 가정할 경우)
df1['인입횟수_ARS_R6M_num'] = (
    df1['인입횟수_ARS_R6M']
    .str.extract(r'(\d+)')      # 숫자만 추출
    .astype(float)              # 숫자형으로 변환
)

# 정제된 컬럼으로 분석
total_count = len(df1)
count_1_to_9 = ((df1['인입횟수_ARS_R6M_num'] >= 1) & (df1['인입횟수_ARS_R6M_num'] < 10)).sum()
count_10_or_more = (df1['인입횟수_ARS_R6M_num'] >= 10).sum()

ratio_1_to_9 = count_1_to_9 / total_count
ratio_10_or_more = count_10_or_more / total_count
ratio_diff = ratio_1_to_9 - ratio_10_or_more

print(f"✅ 1~9회 ARS 이용 비율: {ratio_1_to_9:.2%}")
print(f"✅ 10회 이상 ARS 이용 비율: {ratio_10_or_more:.2%}")
print(f"📊 두 비율 차이: {ratio_diff:.2%}")
```

- ARS를 6개월 내에 이용한 고객 중 ARS를 10회 이상 이용한 비율이 2.96%

### 3_이용메뉴건수_ARS_R6M
- 고객이 최근 6개월간 ARS를 통해 이용한 기능/서비스 메뉴의 총 건수

```python
df1['이용메뉴건수_ARS_R6M'].value_counts()
```

```python
# 주어진 값
data = {
    'label': ['1회 이상', '10회 이상', '20회 이상', '30회 이상'],
    'count': [2313787, 70027, 6014, 10172]
}

df = pd.DataFrame(data)

# 전체 합계
total = df['count'].sum()

# 비율 계산
df['ratio'] = df['count'] / total
df['ratio_percent'] = (df['ratio'] * 100).round(2).astype(str) + '%'

# 출력
print(df)
```

```python
# ARS 인입 1회당 평균 메뉴 수
# 숫자만 추출해서 새 컬럼 만들기 (문자열에서 숫자만 뽑음)
df1['인입횟수_ARS_R6M_num'] = df1['인입횟수_ARS_R6M'].str.extract(r'(\d+)').astype(int)

# 이용메뉴건수도 같은 방식이라면 숫자만 뽑아야 함
df1['이용메뉴건수_ARS_R6M_num'] = df1['이용메뉴건수_ARS_R6M'].str.extract(r'(\d+)').astype(int)

# 이제 숫자로 된 컬럼으로 평균 메뉴 수 계산 후 정수 변환
df1['ARS_인입당_메뉴수'] = (df1['이용메뉴건수_ARS_R6M_num'] / (df1['인입횟수_ARS_R6M_num'] + 1e-5)).round().astype(int)

# 전체 고객 대상 계산
mean_val = df1['ARS_인입당_메뉴수'].mean()
min_val = df1['ARS_인입당_메뉴수'].min()
max_val = df1['ARS_인입당_메뉴수'].max()

print(f"ARS 인입 1회당 평균 메뉴 수: {mean_val:.2f}")
print(f"ARS 인입 1회당 최소 메뉴 수: {min_val}")
print(f"ARS 인입 1회당 최대 메뉴 수: {max_val}")
```

```python
# ARS_인입당_메뉴수가 최대값인 사람
max_val = df1['ARS_인입당_메뉴수'].max()
max_rows = df1[df1['ARS_인입당_메뉴수'] == max_val]

print(f"최대 ARS 인입당 메뉴수: {max_val}인 고객 정보:")
display (max_rows)
```

- 6개월 동안 ARS를 1회 이상 이용한 고객 중 메뉴를 30개 이상 누른 사람은 **14명** 이다.

### 4_인입일수_ARS_R6M

```python
# 인입일수_ARS_R6M의 도수분포를 확인한다.
df1['인입일수_ARS_R6M'].value_counts()
```

```python
# 인입일수가 0인 경우에는 계산할 수 없기 때문에 NaN으로 처리
# 그 외에는 인입일수 기준으로 하루 평균 인입횟수와 메뉴 이용 수를 계산

df1['ARS_일당_인입횟수'] = np.where(
    df1['인입일수_ARS_R6M'] == 0,                             # 인입일수가 0이면
    np.nan,                                                  # NaN 처리
    df1['인입횟수_ARS_R6M_num'] / df1['인입일수_ARS_R6M']     # 아니면 인입횟수 ÷ 인입일수
)

df1['ARS_일당_메뉴수'] = np.where(
    df1['인입일수_ARS_R6M'] == 0,                             # 인입일수가 0이면
    np.nan,                                                  # NaN 처리
    df1['이용메뉴건수_ARS_R6M_num'] / df1['인입일수_ARS_R6M'] # 아니면 메뉴건수 ÷ 인입일수
)

# 📊 계산된 컬럼들의 요약 통계 출력 (평균, 중앙값, 최소/최대 등)
display(df1[['ARS_일당_인입횟수', 'ARS_일당_메뉴수']].describe())
```

- **평균이 0.69, 0.77**: 하루에 한 번 전화(또는 메뉴 이용)가 안 되는 경우도 많고, 보통 하루에 0.5~1회 정도 인입과 메뉴 이용이 있다.
- **최소값 0.08~0.11**: 일당 평균 전화, 메뉴 이용이 거의 없는 수준 (예: 6개월 중 12~9일 정도만 ARS 이용)
- **최대값 2.5, 10**: 어떤 고객은 하루에 2~10번까지 전화하거나 메뉴를 누른 적도 있다는 뜻으로, 이런 고객은 집중적으로 ARS를 이용함

```python
# 각각의 최대값 구하기
max_call = df1['ARS_일당_인입횟수'].max()
max_menu = df1['ARS_일당_메뉴수'].max()

# 최대값 가진 고객 추출
max_call_users = df1[df1['ARS_일당_인입횟수'] == max_call]
max_menu_users = df1[df1['ARS_일당_메뉴수'] == max_menu]

# 결과 확인
display("ARS_일당_인입횟수 최대값 고객 정보:")
display(max_call_users)

display("ARS_일당_메뉴수 최대값 고객 정보:")
display(max_menu_users)
```

- 6개월 동안 ARS를 가장 다양한 일자에 이용한 고객은 총 2명이며, 이들은 ARS를 4일에 걸쳐 이용하였다.
- 하루 동안 가장 많은 메뉴를 이용한 고객은 총 10개 이상의 메뉴를 사용하였다.

- 현금서비스 이용이 불가능한 회원의 수는 44944 (전체의 약 11%에 해당하는 수)

```python
# 블랙리스트인 회원 중 현금 서비스 이용이 가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(cond1 & cond2)
```

```python
# 블랙리스트가 아닌 회원 중 현금 서비스 이용이 가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(~cond1 & cond2)
```

```python
# 블랙리스트인 회원 중 현급 서비스 이용이 불가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(cond1 & ~cond2)
```

```python
# 블랙리스트가 아닌 회원 중 현급 서비스 이용이 불가능한 회원이 있는지 확인한다.
cond1 = df1['회원여부_이용가능'] == 0
cond2 = df1['회원여부_이용가능_CA'] == 1
sum(~cond1 & ~cond2)
```
