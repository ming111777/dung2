```python
# 기본
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import gc

# 경고 뜨지 않게 설정
import warnings
warnings.filterwarnings('ignore')

# 그래프 설정
sns.set()

# 그래프 기본 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# plt.rcParams['font.family'] = 'AppleGothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False

# 결측치 시각화를 위한 라이브러리
import missingno
```

```python
# parquet 파일 데이터를 읽어온다.
df1 = pd.read_parquet('open/train/7.마케팅정보/2018_train_마케팅정보.parquet')
df1
```

### 결측치 찾기

```python
# 결측치가 있는 컬럼을 확인한다.

na_box = []

for idx, val in df1.isnull().sum().items() :
    print(f'{idx}의 결측치 : {val}')
    if val > 0 :
        na_box.append(idx)
```

```python
# 결측치 합계
na_total = df1[na_box].isnull().sum().sum()

print (f'결측치 있는 행: {na_box}')
print(f'전체 결측치 총합: {na_total}개')
```

```python
# 결측치가 있는 칼럼의 결측 비율을 확인한다.
df1[na_box].isna().sum() / df1.shape[0]
```

### 이상치 찾기

```python
import pandas as pd

mixed_cols = []

for col in df1.columns:
    # 결측치는 제외하고, 문자열이 하나라도 포함되어 있는지 확인
    sample = df1[col].dropna().astype(str).head(100)  # 샘플 100개만 검사
    # 숫자만으로 이루어진 값인지 체크
    has_non_numeric = sample.apply(lambda x: not x.isdigit()).any()
    if has_non_numeric:
        mixed_cols.append(col)

print("숫자와 문자가 섞여 있을 가능성이 있는 컬럼:")
for c in mixed_cols:
    print(c)
```

```python
import re

def keep_only_numbers(x):
    # 결측치는 그대로 유지
    if pd.isna(x):
        return np.nan
    # 문자열로 변환 후 숫자만 추출
    s = str(x)
    numbers = re.findall(r'\d+', s)  # 연속된 숫자 덩어리 추출
    if numbers:
        # 숫자 덩어리를 모두 이어붙여 정수형으로 변환
        return int(''.join(numbers))
    else:
        # 숫자가 전혀 없으면 NaN 반환
        return np.nan

# 숫자+문자 섞인 컬럼 리스트
cols_to_clean = [
    '캠페인접촉건수_R12M',
    '캠페인접촉일수_R12M'
]

# 각 컬럼에 대해 숫자만 남기도록 변환 수행
for col in cols_to_clean:
    df1[col] = df1[col].apply(keep_only_numbers)
```

```python
# 숫자가 들어간 컬럼만 선택
numeric_cols = df1.select_dtypes(include=['number']).columns

# 각 컬럼의 값별 개수 전체 출력
for col in numeric_cols:
    print(f"\n🔢 [컬럼: {col}] 값별 빈도수 (전체)")
    print(df1[col].value_counts(dropna=False).sort_index())
```

```python
# ID 제외한 컬럼 선택
num_cols = df1.drop(columns=['ID']).select_dtypes(include='number').columns

outlier_info = []

for col in num_cols:
    # 이상치 기준 계산 (IQR 방식)
    q1 = df1[col].quantile(0.25)
    q3 = df1[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    
    # 이상치 개수 계산
    outlier_count = df1[(df1[col] < lower) | (df1[col] > upper)].shape[0]
    
    if outlier_count > 0:
        outlier_info.append({
            '컬럼명': col,
            '이상치 수': outlier_count,
            '하한값': lower,
            '상한값': upper
        })

# 결과 정리
outlier_df = pd.DataFrame(outlier_info).sort_values(by='이상치 수', ascending=False).reset_index(drop=True)

print("📌 이상치가 있는 컬럼:")
display(outlier_df)
```

```python
import math

# 이상치가 있는 컬럼명 리스트만 추출 (위에서 만든 outlier_df 활용)
cols_with_outliers = outlier_df['컬럼명'].tolist()

# 1행에 5개씩 배치
n_cols = 5
n_rows = math.ceil(len(cols_with_outliers) / n_cols)

# 플롯 크기 설정
plt.figure(figsize=(n_cols * 4, n_rows * 4))

# 박스플롯 그리기
for i, col in enumerate(cols_with_outliers):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.boxplot(data=df1, y=col)
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
import pandas as pd

# 1) ID, OS구분코드, 그리고 날짜형(숫자형으로 저장된) 컬럼을 모두 제외한 수치형 컬럼명 리스트 생성
date_cols = ['기준년월']  

num_cols = (
    df1
    .drop(columns=['ID'] + date_cols)     # 제외할 컬럼들 지정
    .select_dtypes(include='number')                   # 숫자형 타입만 선택
    .columns
)

# 2) 이상치 통계를 담을 리스트 초기화
outlier_stats = []

for col in num_cols:
    # 1사분위(Q1)와 3사분위(Q3)를 계산
    Q1, Q3 = df1[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    
    # 이상치 판정 기준: 하한(lower), 상한(upper)
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # 이상치 마스크, 개수, 비율 계산
    mask = (df1[col] < lower) | (df1[col] > upper)
    count = mask.sum()
    pct   = mask.mean() * 100
    
    # 이상치가 1개 이상인 경우에만 저장
    if count > 0:
        outlier_stats.append((col, count, pct))

# 3) 결과를 DataFrame으로 변환하고, 이상치 비율 기준 내림차순 정렬 후 인덱스 리셋
outlier_df = (
    pd.DataFrame(outlier_stats, columns=['컬럼명', '이상치 개수', '이상치 비율(%)'])
      .sort_values(by='이상치 비율(%)', ascending=False)
      .reset_index(drop=True)
)

# 4) pandas 출력 옵션 변경: 모든 행·열을 표시
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# 5) 결과 출력
print("📌 수치형(날짜형 제외) 컬럼별 이상치 통계:")
display(outlier_df)
```
