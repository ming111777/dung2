### 인사이트 및 제안
- 두 모델 모두 **불만제기후경과월_R12M**와 **OS구분코드_enc**를 최우선으로 봤습니다.  
- XGBoost는 **앱 이용량**, **홈페이지 금융 활동**을, CatBoost는 **PC 방문 기간**, **ARS 인입 기간**을 추가로 중시했습니다.  
- Eta² 분석에서도 **불만제기후경과월_R12M**가 가장 높아, 전 채널 지표 중 ‘불만 이력의 최신성’이 세그먼트 분류에 핵심 변수임을 뒷받침합니다.
