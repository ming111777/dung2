###  프로젝트 세팅

```python
# 학습이 완료된 모델을 저장할 파일 이름
best_model_path = 'model/best_model_classification.dat'

# 예측 결과를 저장할 파일 이름
prediction_path = 'model/classification_prediction.csv'
```

### 저장한 모델 객체 등을 복원한다.

```python
# pickle 파일에서 저장된 모델, 인코더, 스케일러를 순서대로 불러온다.
with open(best_model_path, 'rb') as fp:
    model1 = pickle.load(fp)     # 학습된 모델
    encoder1 = pickle.load(fp)   # 범주형 인코더 (예: OneHotEncoder, LabelEncoder 등)
    scaler1 = pickle.load(fp)    # 스케일러 (예: StandardScaler, MinMaxScaler 등)

# 불러온 객체들을 출력해 확인한다.
display(model1)
display(encoder1)
display(scaler1)
```

```python
# 일부러 오류가 나는 코드를 넣어준다.
# 10 / 0
```

### 예측할 데이터를 준비한다.
- 모든 준비가 완료되면 입력데이터는 **test_X** 라는 변수에 담아준다.

```python
# 예측할 데이터를 읽어온다.
df1 = pd.read_csv('data/breast_cancer_new.csv')
df1
```

```python
# 입력 데이터에 대한 표준화
scaled_data = scaler1.transform(df1)
scaled_data
```

```python
# 입력데이터를 test_X 변수에 담아준다.
test_X = scaled_data
```

### 예측하고 저장한다.

```python
# 예측한다.
y_pred = model1.predict(test_X)
y_pred
```

```python
# 결과데이터를 복원한다.
result = encoder1.inverse_transform(y_pred)
result
```

```python
# 예측 결과를 원본 데이터프레임(df1)에 'target' 컬럼으로 추가한다.
df1['target'] = result

# 결과가 포함된 데이터프레임을 CSV 파일로 저장한다.
df1.to_csv(prediction_path, index=False, encoding='utf-8-sig')
print('저장완료')
```
